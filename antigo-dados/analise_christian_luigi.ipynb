##Preprocessamento

```python
def column_creator(data):
  col = re.sub('"', "", data.columns[0])
  x = col.split('|')
  new = data[data.columns[0]].str.split('|', n=len(x), expand=True)

  d = {}
  for i in range(len(x)):
    d[i] = str(x[i])

  new.rename(columns = d, inplace = True)
  return new
```

```python
column_creator(df_rides)
```

```python
column_creator(df_users)
```

##Convertendo para arquivo CSV

```python
def csv_creator(data, name):
  dados = data.to_csv(index=False)
  arquivo = open(name, 'w')
  arquivo.write(dados)
  arquivo.close()
```

```python
csv_creator(new_users, 'parcial_users.csv')
```

```python
csv_creator(new_rides, 'parcial_rides.csv')
```

```python
csv_creator(new_ride_user, 'parcial_ride_user.csv')
```
