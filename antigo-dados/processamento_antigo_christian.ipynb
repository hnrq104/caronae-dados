#Processamento 2 - dados parcialmente tratados -> dados tratados

```python
parcial = pd.read_csv('dados_user_antigo_parcial2.csv')
```

```python
parcial
```

##Graduate

```python
import matplotlib.pyplot as plt
```

```python
gnames = []
for g in parcial['graduate']:
  if g in gnames:
    continue
  else:
    gnames.append(g)
print(gnames)
print(len(gnames))
print(len(parcial['graduate']))
```

```python
count_list = np.zeros(len(gnames), dtype=np.int64)
print(count_list)
for g in parcial['graduate']:
  if g == 'Graduação':
    count_list[0] += 1
  if g == 'Servidor':
    count_list[1] += 1
  if g == 'Doutorado':
    count_list[2] += 1
  if g == 'Mestrado':
    count_list[3] += 1
  if g == 'Mestrado Profissional':
    count_list[4] += 1
  if g == 'Especialização':
    count_list[5] += 1
  if g == 'Extensão':
    count_list[6] += 1
  if g == 'Residência':
    count_list[7] += 1
  if g == '2':
    count_list[8] += 1
print(count_list)
soma = 0
for c in count_list:
  soma += c
print(soma)

dic = {}
i = 0
for g in gnames:
  dic[g] = count_list[i]
  i += 1
print(dic)
```

```python
fig1, ax1 = plt.subplots(figsize=(10,8))

ax1.pie(count_list, labels=gnames, autopct='%1.1f%%', startangle=90)

ax1.axis('equal')
plt.show()
```
